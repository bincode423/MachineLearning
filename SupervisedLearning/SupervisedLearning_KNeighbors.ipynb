```python
#모듈 호출
from sklearn.neighbors import KNeighborsClassifier #k-최근접 이웃 모듈 호출
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt # 파이썬에서 대표적인 데이터 시각화 모듈
import numpy as np

#로그
print("모듈 호출이 완료되었습니다.")
```

```python
#도미 데이터 분류해보기
bream_length = [25.4, 26.3, 26.5, 29.0, 29.7, 29.7,
                30.0, 30.0, 30.7, 31.0, 31.0, 31.5,
                32.0, 32.0, 32.0, 33.0, 33.0, 33.5,
                33.5, 34.0, 34.0, 34.5, 35.0, 35.0,
                35.0, 35.0, 36.0, 36.0, 37.0, 38.5,
                38.5, 39.5, 40.0, 41.0 , 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0,
                450.0, 500.0, 390.0, 450.0, 500.0,
                475.0, 500.0, 500.0, 340.0, 600.0,
                600.0, 700.0, 700.0, 610.0, 650.0,
                575.0, 685.0, 620.0, 680.0, 700.0,
                725.0, 720.0, 714.0, 850.0, 1000.0,
                920.0, 955.0, 925.0, 975.0, 950.0]
print(f"도미 데이터가 준비되었습니다.")
print(f"길이 데이터 {len(bream_length)}개 / 무게 데이터 {len(bream_weight)}개")

#빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
print(f"빙어 데이터가 준비되었습니다.")
print(f"길이 데이터 {len(smelt_length)}개 / 무게 데이터 {len(smelt_weight)}개")
```

```python
#도미 데이터 시각화
plt.scatter(bream_length, bream_weight) # 데이터 시각화하기 - 산점도
plt.xlabel("length") #x축에 놓을 데이터 설정
plt.ylabel("weight") #y축에 놓을 데이터 설정
plt.show()
```

```python
#빙어 데이터 시각화
plt.scatter(smelt_length, smelt_weight) # 데이터 시각화하기 - 산점도
plt.xlabel("length") #x축에 놓을 데이터 설정
plt.ylabel("weight") #y축에 놓을 데이터 설정
plt.show()
```

```python
#두 데이터를 동시에 보여주기
plt.scatter(bream_length, bream_weight) # 데이터 시각화하기 - 산점도
plt.scatter(smelt_length, smelt_weight) # 데이터 시각화하기 - 산점도
plt.xlabel("length") #x축에 놓을 데이터 설정
plt.ylabel("weight") #y축에 놓을 데이터 설정
plt.show()
```

```python
#도미와 빙어 데이터를 합치기
length = bream_length+smelt_length
weight = bream_weight+smelt_weight

#길이와 무게 데이터를 합치기
#2차원 배열로 변환
fish_data = np.column_stack((length, weight))

#정답 데이터 준비
fish_target = np.concatenate((np.ones(len(bream_length)), np.zeros(len(smelt_length))))

#준비된 데이터 출력
print("준비된 2차원 배열")
print(fish_data)
print("준비된 정답 데이터")
print(">> ", end="")
print(fish_target)
```

```python
#k-최근접 이웃
#'k-최근접 이웃'이란 가까운 데이터 중 가장 많은 것을 채택하여 분류함
kn = KNeighborsClassifier(n_neighbors=15) #객체 생성 #n_neighbors란 몇 개의 가까운 데이터를 확인할지 정하는 것
kn.fit(fish_data, fish_target) #학습시키기
print(round(float(kn.score(fish_data, fish_target))*100, 2), "%") #정확도 채점하기

#데이터가 어떤 생선인지 예측하기
print(kn.predict([[30,600]])) # 도미 
print(kn.predict([[18,5]]))  #빙어

####문제점
# 이미 답을 다 줘 놓고 채점을 함
# 즉 머신러닝의 의미가 없음
```

```python
#랜덤하게 섞어서 샘플 데이터와 학습 데이터 나누기
train_input, test_input, train_target, test_target = \
    train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)

#'k-최근접 이웃' 두 번째 모델
kn = KNeighborsClassifier() #객체 생성
kn.fit(train_input, train_target) #학습시키기
print(round(float(kn.score(test_input, test_target))*100, 2), "%") #정확도 채점하기
```

```python
#수상한 도미
strange = [25,150]
print(kn.predict([strange])) #이것은 도미이지만, 도미로 분류되지 않는다

distances,indexes = kn.kneighbors([strange]) #수상한 도미의 이웃

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(strange[0],strange[1],marker="^") #수상한 도미
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker="D") 

plt.xlabel("length") #x축에 놓을 데이터 설정
plt.ylabel("weight") #y축에 놓을 데이터 설정
plt.show()

#이렇게 보면 이해가 안 됨
```

```python
#표준 점수로 바꾸기
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)

train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std
```

```python
#'k-최근접 이웃' 세 번째 모델
kn = KNeighborsClassifier() #객체 생성
kn.fit(train_scaled, train_target) #학습시키기
print(round(float(kn.score(test_scaled, test_target))*100, 2), "%") #정확도 채점하기
```

```python
#다시 수상한 도미
strange = ([25,150] - mean)/ std

print(kn.predict([strange])) #이것은 도미다.
distances,indexes = kn.kneighbors([strange])


plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(strange[0],strange[1],marker="^") #수상한 도미
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], marker="D") # 

plt.xlabel("length") #x축에 놓을 데이터 설정
plt.ylabel("weight") #y축에 놓을 데이터 설정
plt.show()
```
